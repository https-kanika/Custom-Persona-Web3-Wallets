```python
# ✅ Step 1: Install Moralis if not installed
!pip install moralis

# ✅ Step 2: Imports
import pandas as pd
from moralis import evm_api
from tqdm import tqdm

# ✅ Step 3: Config
api_key = "eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.eyJub25jZSI6IjQ2ODIxMDJkLTI2NWQtNGVmNi05ZTA1LTc0ZmUxZDY5YjlmMyIsIm9yZ0lkIjoiNDQ5NzQzIiwidXNlcklkIjoiNDYyNzQzIiwidHlwZUlkIjoiZWI1NDE2MjctNDg5YS00NTJhLTgyYmYtMTJiZDIwYTMwOTYxIiwidHlwZSI6IlBST0pFQ1QiLCJpYXQiOjE3NDg0NjQ2NjAsImV4cCI6NDkwNDIyNDY2MH0.M5yiljp0HmqgtHSymp2rUzzGyWfMOM152-mU4za0zDw"  # Replace with your Moralis key
input_csv_path = "/content/drive/MyDrive/web3_kgenX_new/wallets.csv"
output_csv_path = "/content/drive/MyDrive/web3_kgenX_new/nft_collections_cleaned.csv"
chain = "eth"

# ✅ Step 4: Load Wallet IDs
df_wallets = pd.read_csv(input_csv_path)
wallets = df_wallets['wallet_ID'].dropna().unique()

# ✅ Step 5: Fetch and filter NFT collections
cleaned_collections = []

for wallet in tqdm(wallets, desc="Processing wallets"):
    try:
        params = {
            "chain": chain,
            "address": wallet
        }

        result = evm_api.nft.get_wallet_nft_collections(
            api_key=api_key,
            params=params
        )

        collections = result.get("result", [])
        if isinstance(collections, dict):  # If result is a single dict
            collections = [collections]

        for col in collections:
            filtered = {
                "token_address": col.get("token_address", ""),
                "contract_type": col.get("contract_type", ""),
                "name": col.get("name", ""),
                "verified_collection": col.get("verified_collection", ""),
                "count": col.get("count", 0)
            }
            cleaned_collections.append(filtered)

    except Exception as e:
        print(f"❌ Error with wallet {wallet}: {e}")
        continue

# ✅ Step 6: Save final cleaned DataFrame
df_cleaned = pd.DataFrame(cleaned_collections)
df_cleaned.to_csv(output_csv_path, index=False)
print(f"✅ Saved cleaned NFT collection data to: {output_csv_path}")
```
